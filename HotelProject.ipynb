# Chart - 3
Question no. 3 Which Type Of hotel have Highest revenue or ADR[ City hotels or Resort Hotels ]

````python
# Group by Hotel and calculate the average ADR
avg_adr = df1.groupby('hotel')['adr'].mean().reset_index()

# Display the average ADR values
print(avg_adr)

# Plot the comparison graph with the updated method
plt.figure(figsize=(8, 6))
sns.barplot(data=avg_adr, x='hotel', y='adr', hue='hotel', palette=['blue', 'green'], dodge=False)
plt.xlabel('Hotel Type')
plt.ylabel('Average ADR')
plt.title('Comparison of Average ADR: City Hotels vs Resort Hotels')
plt.legend([], [], frameon=False)  # Remove legend as it's redundant
plt.show()

````

# 1.Why did you pick the specific Chart ?

The bar chart was chosen because it is effective for comparing categorical data, such as the Average Daily Rate (ADR) of City Hotels vs. Resort Hotels. It provides a clear and direct visualization of how each hotel type performs, making it easy to see which hotel type generates higher revenue per room.

# 2.What is/are the insight(s) found from the chart?

The chart shows that **City Hotels** have a higher Average Daily Rate (ADR) of **105.50** compared to **Resort Hotels**, which have an ADR of **94.98**.  

This indicates that City Hotels generate more revenue per room, possibly due to:  
- Strong demand from business travelers and urban tourists.  
- Higher room rates in central locations where guests may be willing to pay a premium for convenience.  
- Consistent occupancy rates, particularly during weekdays when business travel is high.  

This insight suggests that focusing on enhancing services and experiences in City Hotels could further boost profitability.

````python
# Chart - 8 visualization code(agent made the most number of bookings)
top_bookings_by_agent = df1['agent'].value_counts().reset_index().rename(columns={'index':'agent','agent':'num_of_bookings'})[:10]
top_bookings_by_agent
````

````python


# Sort data by count in descending order
top_bookings_by_agent = top_bookings_by_agent.sort_values(by="count", ascending=False)

# Set figure size
plt.figure(figsize=(11, 6))

# Create barplot with different colors for each bar
sns.barplot(x="num_of_bookings",
            y="count",
            data=top_bookings_by_agent,
            palette=sns.color_palette("hsv", len(top_bookings_by_agent)),  # Assigns unique colors
            order=top_bookings_by_agent["num_of_bookings"])  # Ensure correct order

# Add labels and title
plt.xlabel("Agent ID", fontsize=14)
plt.ylabel("Number of Bookings", fontsize=14)
plt.title("Most Bookings by the Agent", fontsize=18)

# Rotate x-axis labels for better visibility (if needed)
plt.xticks(rotation=45)

# Display the plot
plt.show()

````
